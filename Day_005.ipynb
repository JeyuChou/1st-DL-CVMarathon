Day 005_1 --> creating a dataframe

```
import pandas as pd
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import io
import os
```

```
cities = ['Austin','Dallas','Houston']
ranking = [np.random.randint(4) for i in range(3)]
population = [np.random.randint(10000,100000000)for i in range(3)]
```

```
result = zip(cities,ranking,population)
data = list(result)
```

```
df = pd.DataFrame(data,index= None,columns=['cities','ranking','population'])
```

```
df.groupby(by='cities')['population'].mean()
```
